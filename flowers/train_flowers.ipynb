```python
import ultralytics
import yaml
print(ultralytics.__file__)
```

```python
default_yaml_path = '/usr/local/lib/python3.10/dist-packages/ultralytics/cfg/default.yaml'

with open(default_yaml_path, 'r') as f:
    default_yaml = f.read()

print(default_yaml)
```

```python
with open(default_yaml_path, 'r') as file:
    config = yaml.safe_load(file)


config['degrees'] = 10
config['scale'] = 0.8
config['translate'] = 0.3
```

```python
with open(default_yaml_path, 'w') as file:
    yaml.dump(config, file)
```
